```python
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# 그래프 생성
G = nx.Graph()

edges = [
    ("소테", "유림닭도리탕", 37),
    ("유림닭도리탕", "맛있는 칼국수", 14),
    ("맛있는 칼국수", "오몬자", 11),
    ("오몬자", "남산광어", 37),
    ("남산광어", "와이인", 29),
    ("와이인", "효뜨", 21),
    ("효뜨", "문화식당", 36),
    ("문화식당", "스페이스", 22),
    ("스페이스", "르부아", 18),
    ("르부아", "오몬자", 37)
]

G.add_weighted_edges_from(edges)

# 노드 위치
pos = nx.spring_layout(G, seed=42)

# DFS 순서 저장
visited = []
dfs_edges = []

def dfs(node):
    visited.append(node)

    for neighbor in G.neighbors(node):
        if neighbor not in visited:
            dfs_edges.append((node, neighbor))
            dfs(neighbor)

# 시작 노드
start_node = list(G.nodes)[0]
dfs(start_node)

# 애니메이션
fig, ax = plt.subplots(figsize=(10, 8))

def update(frame):
    ax.clear()

    current_nodes = visited[:frame+1]
    current_edges = dfs_edges[:frame]

    node_colors = []

    for node in G.nodes():
        if node in current_nodes:
            node_colors.append("orange")
        else:
            node_colors.append("lightgray")

    nx.draw(
        G,
        pos,
        with_labels=True,
        node_color=node_colors,
        node_size=2000,
        font_family="Malgun Gothic",
        ax=ax
    )

    nx.draw_networkx_edges(
        G,
        pos,
        edgelist=current_edges,
        width=3,
        edge_color="red",
        ax=ax
    )

    labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

    ax.set_title(f"DFS Step {frame+1}")

ani = FuncAnimation(
    fig,
    update,
    frames=len(visited),
    interval=1000,
    repeat=False
)

ani.save("dfs.mp4", writer="ffmpeg")

plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from collections import deque

# 그래프 생성
G = nx.Graph()

edges = [
    ("소테", "유림닭도리탕", 37),
    ("유림닭도리탕", "맛있는 칼국수", 14),
    ("맛있는 칼국수", "오몬자", 11),
    ("오몬자", "남산광어", 37),
    ("남산광어", "와이인", 29),
    ("와이인", "효뜨", 21),
    ("효뜨", "문화식당", 36),
    ("문화식당", "스페이스", 22),
    ("스페이스", "르부아", 18),
    ("르부아", "오몬자", 37)
]

G.add_weighted_edges_from(edges)

# 노드 위치
pos = nx.spring_layout(G, seed=42)

# BFS 순서 저장
visited = []
bfs_edges = []

def bfs(start):
    queue = deque([start])
    visited_set = set([start])

    while queue:
        node = queue.popleft()
        visited.append(node)

        for neighbor in G.neighbors(node):
            if neighbor not in visited_set:
                visited_set.add(neighbor)
                queue.append(neighbor)
                bfs_edges.append((node, neighbor))

# 시작 노드
start_node = list(G.nodes)[0]
bfs(start_node)

# 애니메이션
fig, ax = plt.subplots(figsize=(10, 8))

def update(frame):
    ax.clear()

    current_nodes = visited[:frame+1]
    current_edges = bfs_edges[:frame]

    node_colors = []

    for node in G.nodes():
        if node in current_nodes:
            node_colors.append("skyblue")
        else:
            node_colors.append("lightgray")

    nx.draw(
        G,
        pos,
        with_labels=True,
        node_color=node_colors,
        node_size=2000,
        font_family="Malgun Gothic",
        ax=ax
    )

    nx.draw_networkx_edges(
        G,
        pos,
        edgelist=current_edges,
        width=3,
        edge_color="blue",
        ax=ax
    )

    labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

    ax.set_title(f"BFS Step {frame+1}")

ani = FuncAnimation(
    fig,
    update,
    frames=len(visited),
    interval=1000,
    repeat=False
)

ani.save("bfs.mp4", writer="ffmpeg")

plt.show()
```
